```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
covid = pd.read_csv("https://opendata.arcgis.com/datasets/6811473e86fe4e2cb6af10d54c15ecee_0.csv")
```

```python
covid = covid.sort_values(by=['Date'])
```

```python
del covid['DthCOVID19']
del covid['Outbreaks']
```

```python
covid = covid.dropna()
covid
```

```python
covid_change = pd.DataFrame([covid['Date'], covid['Tested'].diff(), covid['Cases'].diff(), covid['Deaths'].diff()]).T
```

```python
covid_change = covid_change.dropna()
```

```python
covid_change['PositiveRate'] = covid_change['Cases'] / covid_change['Tested'] * 100
```

```python
positive_rate = go.Figure(data=[go.Bar(x=covid_change['Date'], y=covid_change['PositiveRate'], name='% Tests that are positive')])
positive_rate.update_yaxes(title_text="<b>Percent</b>", showgrid=False)
positive_rate.update_xaxes(title_text="<b>Date</b>", showgrid=False)
positive_rate.show()
```

```python
fig = make_subplots(specs=[[{"secondary_y": True}]])
fig.add_trace(go.Bar(x=covid_change['Date'], y=covid_change['Tested'], name='Tests'), secondary_y=False)
fig.add_trace(go.Scatter(x=covid_change['Date'], y=covid_change['Cases'], name='Cases'), secondary_y=True)
fig.update_yaxes(title_text="<b>Tests that Day</b>", secondary_y=False, showgrid=False)
fig.update_yaxes(title_text="<b>Positive Tests that Day</b>", secondary_y=True, showgrid=False)
fig.update_xaxes(title_text="<b>Date</b>", showgrid=False)
fig
```

```python
deaths = go.Figure(data=[go.Bar(x=covid_change['Date'], y=covid_change['Deaths'], name='Deaths')])
deaths.update_yaxes(title_text="<b>Deaths that Day</b>", showgrid=False)
deaths.update_xaxes(title_text="<b>Date</b>", showgrid=False)
deaths.show()
```
